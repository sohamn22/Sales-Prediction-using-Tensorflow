```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
df.head()
```

```python
df.drop('customerID',axis='columns', inplace=True)
```

```python
df.dtypes
```

```python
df.TotalCharges.values
```

```python
df['TotalCharges']=pd.to_numeric(df.TotalCharges, errors='coerce')
#Converting Total Charges from string to int and ignoring errors and null spaces
```

```python
mean_total_charges=df['TotalCharges'].mean()
```

```python
mean_total_charges
```

```python
df['TotalCharges'].fillna(mean_total_charges,inplace=False)
```

```python
df.TotalCharges.dtypes
```

```python
tenure_churn_no=df[df.Churn=='No'].tenure
tenure_churn_yes=df[df.Churn=='Yes'].tenure
```

```python
plt.xlabel("TENURE")
plt.xlabel("Number of Customers")
plt.title("Customer Churn Prediction")
plt.hist([tenure_churn_yes,tenure_churn_no],color=['green','red'], label=['Churn=Yes', 'Churn=No'])
plt.legend()
```

```python
def print_unique_values(df):
  for column in df:
    if df[column].dtypes=='object':
      print(f'{column}: {df[column].unique()}')
```

```python
print_unique_values(df)
```

```python
df.replace('No internet service','No',inplace=True)
df.replace('No phone service','No',inplace=True)
```

```python
print_unique_values(df)
```

```python
yes_no_columns=['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']

for col in yes_no_columns:
  df[col].replace({'Yes':1,'No':0},inplace=True)
#Replacing unique values like yes or no with 1 and 0
```

```python
print_unique_values(df)
```

```python
for column in df:
  print(f'{column}: {df[column].unique()}')
```

```python
df['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
for column in df:
  print(f'{column}: {df[column].unique()}')
```

```python
df4 = pd.get_dummies(data=df, columns=['InternetService', 'PaymentMethod', 'Contract'])

# Identify dummy variable columns (those newly created by pd.get_dummies)
dummy_columns = df4.columns.difference(df.columns)

# Convert only the dummy variable columns to integers
df4[dummy_columns] = df4[dummy_columns].fillna(0).astype(int)
```

```python
df4.head()
```

```python
for column in df4:
  print(f'{column}: {df4[column].unique()}')
```

```python
df4.dtypes
```

```python
cols_to_scale=['tenure','MonthlyCharges','TotalCharges']
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df4[cols_to_scale]=scaler.fit_transform(df4[cols_to_scale])
```

```python
for column in df4:
  print(f'{column}: {df4[column].unique()}')
```

```python
X= df4.drop('Churn',axis='columns')
y=df4['Churn']
```

```python
X=X.fillna(mean_total_charges)
```

```python
X.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
print(np.any(np.isnan(X_train)))
print(np.any(np.isinf(X_train)))
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model=keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,), activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')

])
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train,epochs=10)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
y_pred[:5]
```

```python
y_predicted = []
for element in y_pred:
    if element > 0.5:
        y_predicted.append(1)
    else:
        y_predicted.append(0)
```

```python
#compare y_predicted and y_test to compare
```

```python
from sklearn.metrics import confusion_matrix , classification_report

print(classification_report(y_test,y_predicted))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
